```python
#The csv files generated by the APIs in the other program code
csvpath = os.path.join("shootings2018.csv")
csvpath2 = os.path.join("officer_outreach.csv")

shootings_data = pd.read_csv(csvpath)
outreach_data = pd.read_csv(csvpath2, encoding="ANSI")
```

```python
#The calculated coordinates of the highest concentration of shootings and outreach
csvpath = os.path.join("danger2018.csv")
csvpath2 = os.path.join("outreach.csv")

danger_center_data = pd.read_csv(csvpath)
outreach_center_data = pd.read_csv(csvpath2)#, encoding="ANSI")
```

```python
# Place map
gmap = gmplot.GoogleMapPlotter(outreach_data["Latitude"].mean(), outreach_data["Longitude"].mean(), 11.5)

arr = []
for i in range(len(shootings_data["Longitude"])):
    arr.append((shootings_data["Latitude"][i],shootings_data["Longitude"][i]))
    incident = shootings_data["INCIDENT_ID"][i]
    gmap.marker(shootings_data['Latitude'][i], shootings_data['Longitude'][i],title=shootings_data['INDIVIDUAL_INJURY_TYPE'][i])

# Scatter points
top_attraction_lats, top_attraction_lons = zip(*arr)
gmap.scatter(top_attraction_lats, top_attraction_lons, 'rgb(200,0,0)', size=200, marker=False)




arr2 = []
for i in range(len(outreach_data["Longitude"])):
    arr2.append((outreach_data["Latitude"][i],outreach_data["Longitude"][i]))

# Scatter points
top_attraction_lats, top_attraction_lons = zip(*arr2)
gmap.scatter(top_attraction_lats, top_attraction_lons, 'rgb(0,0,200)', size=200, marker=False)

danger_lats, danger_lngs = zip(*[
    (danger_center_data.iloc[0]["Latitude"],danger_center_data.iloc[0]["Longitude"]),
    (danger_center_data.iloc[1]["Latitude"],danger_center_data.iloc[1]["Longitude"]),
    (danger_center_data.iloc[2]["Latitude"],danger_center_data.iloc[2]["Longitude"]),
    (danger_center_data.iloc[3]["Latitude"],danger_center_data.iloc[3]["Longitude"]),
    (danger_center_data.iloc[4]["Latitude"],danger_center_data.iloc[4]["Longitude"])
                                 
    #(danger_lat-0.02, danger_lng-0.02),
    #(danger_lat+0.02, danger_lng-0.02),
    #(danger_lat+0.02, danger_lng+0.02),
    #(danger_lat-0.02, danger_lng+0.02),
    #(danger_lat-0.02, danger_lng-0.02)
     ])
#freq_lat2, freq_lng2 = zip(*freq_arr)
#gmap.scatter(danger_lats, danger_lngs, 'cornflowerblue', size=80, marker=False)
gmap.plot(danger_lats, danger_lngs, 'orange', edge_width=10)


outreach_lats, outreach_lngs = zip(*[
    (outreach_center_data.iloc[0]["Latitude"],outreach_center_data.iloc[0]["Longitude"]),
    (outreach_center_data.iloc[1]["Latitude"],outreach_center_data.iloc[1]["Longitude"]),
    (outreach_center_data.iloc[2]["Latitude"],outreach_center_data.iloc[2]["Longitude"]),
    (outreach_center_data.iloc[3]["Latitude"],outreach_center_data.iloc[3]["Longitude"]),
    (outreach_center_data.iloc[4]["Latitude"],outreach_center_data.iloc[4]["Longitude"])
                                 
     ])
#freq_lat2, freq_lng2 = zip(*freq_arr)
#gmap.scatter(danger_lats, danger_lngs, 'cornflowerblue', size=80, marker=False)
gmap.plot(outreach_lats, outreach_lngs, 'cornflowerblue', edge_width=10)


# Marker
#hidden_gem_lat, hidden_gem_lon = outreach_data["Latitude"].mean(), outreach_data["Longitude"].mean()
#gmap.marker(hidden_gem_lat, hidden_gem_lon, 'cornflowerblue')

# Draw
gmap.draw("Combined_Map.html")
```
