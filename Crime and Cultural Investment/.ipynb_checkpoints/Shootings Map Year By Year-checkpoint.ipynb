```python
# Dependencies
import requests
import json
import pandas as pd
import numpy as np
# Google developer API key
from config import gkey
from gmplot import gmplot
```

```python
#The last year included in the analysis or only year included
Yr=2018
#If you want cumulative maps, change this to False
only_year=False
```

```python

#Victim API
url2 = "http://clt-charlotte.opendata.arcgis.com/datasets/2564657b1e9a4883ab388badadbeef6d_5.geojson"
response2 = requests.get(url2)
df_victim = pd.DataFrame(response2.json()['features'])
arr2 = []
for b in df_victim["properties"]:
    arr2.append(b)
df_victim = pd.DataFrame(arr2)
df_victim
```

```python
#This drops an incident that is missing a latitude or a longitude
df_victim = df_victim.dropna(axis=0, subset=['Latitude', 'Longitude'],how='any')
df_victim=df_victim.reset_index(drop=True)
```

```python
#This drops an incident that is outside our year range
for row in range(len(df_victim)):
    if only_year==False:
        if df_victim["YR"][row]>Yr:
            df_victim=df_victim.drop(row)
    else:
        if df_victim["YR"][row]!=Yr:
            df_victim=df_victim.drop(row)
        

df_victim=df_victim.reset_index(drop=True)
df_victim
```

```python
#Finding the shooting with the most "neighbors"-defined as 0.02 degrees east, west, north, and south
#This will give us a area of highest concentration
near=[]
close_ones=[]
#dummy=pd.DataFrame({'Nearby Shooting(by index)':near})
for row in (range(len(df_victim))):
    target_lat=df_victim["Latitude"][row]
    target_lng=df_victim["Longitude"][row]
    for comp_row in (range(len(df_victim))):
        if (abs(df_victim["Latitude"][comp_row]-target_lat)<0.02) & (abs(df_victim["Longitude"][comp_row]-target_lng)<0.02):
            if row!=comp_row:
                close_row=comp_row
                near.append(close_row)
    #print(near)
    #df = pd.DataFrame({'Nearby Shootings': near})
    #df = {'Nearby Shootings': near}
    #df = near
    #df_victim.loc[row,"Nearby Shooting(by index)"]=df
   # a = np.arange(10).reshape(2,5)
    #b = np.arange(10, 20).reshape(2,5)
    close_ones.append(near)
    #dummy.head()
    near=[]
#df_victim.head()
#print(len(close_ones))
#print(len(df_victim))
#print(close_ones[3])
max_neighbors=0
danger_zone_index=""
for x in range(len(close_ones)):
    #print(str(x)+" has: "+str(len(close_ones[x])))
    numb_close=len(close_ones[x])
    if numb_close > max_neighbors:
        max_neighbors=numb_close
        danger_zone_index=x
if max_neighbors!=0:
    print("List #"+str(danger_zone_index)+" is the most dangerous place with " + str(max_neighbors))
    danger_lat=df_victim["Latitude"][danger_zone_index]
    danger_lng=df_victim["Longitude"][danger_zone_index]
else:
    print("No concentration of officer involved shootings in this time period.")

```

```python
# Place map
gmap = gmplot.GoogleMapPlotter(df_victim["Latitude"].mean(), df_victim["Longitude"].mean(), 11.5)

arr = []
for i in range(len(df_victim["Longitude"])):
    arr.append((df_victim["Latitude"][i],df_victim["Longitude"][i]))
    incident = df_victim["INCIDENT_ID"][i]
    #description=df_incident.loc[df_incident["INCIDENT_ID"]==incident,["NARRATIVE"]]
    #try:
    #    gmap.marker(df_victim['Latitude'][i], df_victim['Longitude'][i],
    #                title=df_incident.loc[df_incident["INCIDENT_ID"]==incident,["NARRATIVE"]])
    #gmap.marker(df_victim['Latitude'][i], df_victim['Longitude'][i],title=description)
    gmap.marker(df_victim['Latitude'][i], df_victim['Longitude'][i],title=df_victim['INDIVIDUAL_INJURY_TYPE'][i])

if len(df_victim)!=0:
    # Scatter points
    shots_lats, shots_lons = zip(*arr)
    gmap.scatter(shots_lats, shots_lons, 'rgb(200,0,0)', size=200, marker=False)
else:
    gmap.marker(35.24, -80.82,title=":)")

if max_neighbors !=0:
    #Draw the "danger zone" where there have been the most officer involved shootings
    danger_lats, danger_lngs = zip(*[
        (danger_lat-0.02, danger_lng-0.02),
        (danger_lat+0.02, danger_lng-0.02),
        (danger_lat+0.02, danger_lng+0.02),
        (danger_lat-0.02, danger_lng+0.02),
        (danger_lat-0.02, danger_lng-0.02)
         ])

    gmap.plot(danger_lats, danger_lngs, 'orange', edge_width=10)


# Marker
#This marker icon must be altered manually
KLS_lat, KLS_lng = 35.29573, -80.726731
gmap.marker(KLS_lat, KLS_lng, 'cornflowerblue')

# Draw
if only_year==True:
    gmap.draw("CMPD_Shootings_Map_"+str(Yr)+"exclusive.html")
else:
    gmap.draw("CMPD_Shootings_Map_"+str(Yr)+".html")
```
