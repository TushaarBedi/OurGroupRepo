```python
# Dependencies
import requests
import json
import pandas as pd
# Google developer API key
from config import gkey
from gmplot import gmplot
from pprint import pprint
```

```python
#Officer Outreach Events API
event_url = "http://clt-charlotte.opendata.arcgis.com/datasets/2294932b389048a4bd650fe6d305aafb_13.geojson"
event_response = requests.get(event_url)
df_event = pd.DataFrame(event_response.json()['features'])
arr_e = []
for b in df_event["properties"]:
    arr_e.append(b)
df_event = pd.DataFrame(arr_e)
df_event.head()
```

```python
#This drops an incident that is missing a latitude or a longitude
df_event = df_event.dropna(axis=0, subset=['Location'],how='any')
df_event=df_event.reset_index(drop=True)
len(df_event)
```

```python
#Drops entries that have no addresses. 
df_clean=df_event.copy()
for i in range(len(df_event)):
    if any(x.isdigit() for x in df_event.iloc[i]["Location"])==False:
        df_clean=df_clean.drop([i])
df_clean=df_clean.reset_index(drop=True)
df_clean.head()
```

```python
#If we ever needed to inspect the entries without addresses
df_trouble=df_event.copy()
for i in range(len(df_event)):
    if any(x.isdigit() for x in df_event.iloc[i]["Location"])==True:
        df_trouble=df_trouble.drop([i])
df_trouble=df_trouble.reset_index(drop=True)
df_trouble.head()
```

```python
#This cleans the front half of the location entries. 
#"Saint Michael's 1101 Park Rd" becomes "1101 Park Rd"
for i in range(len(df_clean)):
    to_drop=""
    for x in df_clean.iloc[i]["Location"]:
        if x.isdigit()==False:
            to_drop = to_drop+x
        else:
            break
    new_string=df_clean.iloc[i]["Location"].replace(to_drop,'')
    df_clean.loc[i,"Simple Location"]=new_string
df_clean.head(1)
```

```python
#Create a function to find the end of an address.
def find_address_end(key_word, word):
    return word.rfind(key_word)+len(key_word)
check=["drive","dr","drive("," way"," street"," st"," avenue"," ave"," boulevard"," bl "," bl."," blvd"," road"," rd"," circle"," cir "," cir."," lane "," ln"," highway"," hwy"]
```

```python
#Now that we know the end of the address, we can chop the extra back matter
#So, "1101 Park Rd., third door on the right" becomes "1101 Park Rd.
#Also creates the correct searchable address for Google's API
for i in range(len(df_clean)):
    new_word_loc=0
    found=False
    word=df_clean.iloc[i]["Simple Location"].lower()
    for each in check:
        new_word_loc=find_address_end(each,word)
        if new_word_loc != len(each)-1:
            found = True
            break
    if found:
        new_word=word[0:new_word_loc]
    else:
        new_word=word
    df_clean.loc[i,"Simple Location"]=new_word.replace(" ","+")+",+Charlotte,+NC"
df_clean.head()
```

```python
#Finds the latitude and longitude for the addresses
#Many entries makes this step take a few minutes
base_url="https://maps.googleapis.com/maps/api/geocode/json?address="
key_base="&key="+gkey
for row in range(len(df_clean)):
    problem=False
    response=requests.get(base_url+df_clean.iloc[row]["Simple Location"]+key_base,"Problem").json()

    if response =="Problem":
        
        df_clean.loc[row,"Latitude"]=float('nan')
        df_clean.loc[row,"Longitude"]=float('nan')

    else:
        try:
            df_clean.loc[row,"Latitude"]=response["results"][0]["geometry"]["location"]["lat"]
            df_clean.loc[row,"Longitude"]=response["results"][0]["geometry"]["location"]["lng"]
        except:
            df_clean.loc[row,"Latitude"]=float('nan')
            df_clean.loc[row,"Longitude"]=float('nan')
    
df_clean.head()
```

```python
#This drops an incident that is missing a latitude or a longitude
df_clean = df_clean.dropna(axis=0, subset=['Latitude', 'Longitude'],how='any')
df_clean=df_clean.reset_index(drop=True)
#len(df_clean)
```

```python
#Find the highest concentration of police outreach locations
near=[]
close_ones=[]
#distance is how ever close you consider a "close" neighbor, in degrees lat & lng
distance=0.02
#Creates array of close neighbors' indices
for row in (range(len(df_clean))):
    target_lat=df_clean["Latitude"][row]
    target_lng=df_clean["Longitude"][row]
    for comp_row in (range(len(df_clean))):
        if (abs(df_clean["Latitude"][comp_row]-target_lat)<distance) & (abs(df_clean["Longitude"][comp_row]-target_lng)<distance):
            if row!=comp_row:
                close_row=comp_row
                near.append(close_row)
    close_ones.append(near)
    near=[]

#Now that we know all the neighbors, let's see who has the most
max_neighbors=0
outreach_zone_index=""
for x in range(len(close_ones)):
    numb_close=len(close_ones[x])
    if numb_close > max_neighbors:
        max_neighbors=numb_close
        outreach_zone_index=x
print("List #"+str(outreach_zone_index)+" is the most outreached place with " + str(max_neighbors)+" neighborhood activities")

outreach_lat=df_clean["Latitude"][outreach_zone_index]
outreach_lng=df_clean["Longitude"][outreach_zone_index]
```

```python
# Place map
gmap = gmplot.GoogleMapPlotter(df_clean["Latitude"].mean(), df_clean["Longitude"].mean(), 11.5)

arr = []
for i in range(len(df_clean["Longitude"])):
    arr.append((df_clean["Latitude"][i],df_clean["Longitude"][i]))
    #gmap.marker(df_clean['Latitude'][i], df_clean['Longitude'][i],title=df_clean['Title'][i])


# Scatter points
top_attraction_lats, top_attraction_lons = zip(*arr)
gmap.scatter(top_attraction_lats, top_attraction_lons, 'rgb(0,0,200)', size=200, marker=False)


#Creates a box around the highest concentration of CMPD outreach
outreach_lats, outreach_lngs = zip(*[
    (outreach_lat-distance, outreach_lng-distance),
    (outreach_lat+distance, outreach_lng-distance),
    (outreach_lat+distance, outreach_lng+distance),
    (outreach_lat-distance, outreach_lng+distance),
    (outreach_lat-distance, outreach_lng-distance)
     ])

gmap.plot(outreach_lats, outreach_lngs, 'cornflowerblue', edge_width=10)

# Marker
#hidden_gem_lat, hidden_gem_lon = df_clean["Latitude"].mean(), df_clean["Longitude"].mean()
#gmap.marker(hidden_gem_lat, hidden_gem_lon, 'cornflowerblue')

# Draw
gmap.draw("CMPD_Outreach_Map.html")
```
