```python

#Officer Outreach Events API
event_url = "http://clt-charlotte.opendata.arcgis.com/datasets/2294932b389048a4bd650fe6d305aafb_13.geojson"
event_response = requests.get(event_url)
df_event = pd.DataFrame(event_response.json()['features'])
arr_e = []
for b in df_event["properties"]:
    arr_e.append(b)
df_event = pd.DataFrame(arr_e)
df_event
```
