```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#victim api
url2 = "http://clt-charlotte.opendata.arcgis.com/datasets/2564657b1e9a4883ab388badadbeef6d_5.geojson"
response2 = requests.get(url2)
df = pd.DataFrame(response2.json()['features'])
arr2 = []
for b in df["properties"]:
   arr2.append(b)
df2 = pd.DataFrame(arr2)

df2
```

```python
#victims by gender 
gender_group = df2.groupby('INDIVIDUAL_GENDER')
gender_count = gender_group['INDIVIDUAL_GENDER'].count()
gender_count
gender_df = pd.DataFrame({"Gender Count": gender_count})
gender_df_index = gender_df.reset_index()

gender_df_index
```

```python
# Code for pie chart for victim gender
# Labels for the sections of our pie chart
labels = gender_df_index["INDIVIDUAL_GENDER"]
# The values of each section of the pie chart
sizes = gender_df_index["Gender Count"]

# The colors of each section of the pie chart
colors = ["lightskyblue", "lightcoral"]
# Tells matplotlib to seperate the Suburban and Rural sections from the others
explode = (0, 0.1)
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=200)

# Create Title
plt.title("Percentage of Shooting Victims by Gender for Charlotte Data (2005-2018)", fontsize = 15)
# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

# Saving the Image
#plt.savefig(â€œ../Images/Age Percentage.pngâ€) # saving the graph as a .png image
plt.savefig("Desktop/PythonStuff/Gender Percentage Pie.png")
# Prints our pie chart to the screen
plt.show()
```

```python
#victims by race 
race_group = df2.groupby('INDIVIDUAL_RACE')
race_count = race_group['INDIVIDUAL_RACE'].count()
race_count
race_df = pd.DataFrame({"Race Count": race_count})
race_df_index = race_df.reset_index()

race_df_index
```

```python
# Code for pie chart for victim race
# Labels for the sections of our pie chart
labels = ["Asian", "African American", "Hispanic", "Some Other Race", "Caucasian"]
# The values of each section of the pie chart
sizes = race_df_index["Race Count"]

# The colors of each section of the pie chart
colors = ["lightyellow", "gray", "lightskyblue", "cyan", "lightcoral"]
# Tells matplotlib to seperate the Suburban and Rural sections from the others
explode = (0, 0.1, 0.1, 0.1, 0.1)
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=200)

# Create Title
plt.title("Percentage of Shooting Victims by Race for Charlotte Data (2005-2018)", fontsize = 15)
# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

# Saving the Image
#plt.savefig(â€œ../Images/Age Percentage.pngâ€) # saving the graph as a .png image
plt.savefig("Desktop/PythonStuff/Race Percentage Pie.png")
# Prints our pie chart to the screen
plt.show()
```

```python
#binning
max_age = df2["INDIVIDUAL_AGE"].max()
min_age = df2["INDIVIDUAL_AGE"].min()

# Bins defined by Victimâ€™s Age Group as: <13 = Child, 13 â€“ 17 = â€œTeenâ€, 18 â€“ 22 = â€œCollegeâ€, 
# 23 â€“ 30 = â€œYoung Adultâ€, 31 â€“ 64 = â€œAdultâ€, 65 â€“ 87 = â€œSeniorâ€; pie chart)

bins = [0, 17, 22, 30, 40, 50, 65, max_age]

group_names = ['13-17 = Teen','18-22 = College','23-30 = Young Adult','31-40 = 30s', '41-50 = 40s', '51-65 = 50s - 60s', '65+ = Senior']

# Using the Bin function to split our data by â€œAge Demogrpahicsâ€ into 6 bins
               
df2["Age Demographics"] = pd.cut(df2["INDIVIDUAL_AGE"], bins, labels=group_names)
df2.head() # Display the dataframe
```

```python
age_df2 = df2.groupby("Age Demographics")
age_df2.count()
```

```python
# Counting the shootings by age
age_shooting_count = age_df2["INCIDENT_ID"].count()
# Creating a dataframe for count of shootings by age
age_shooting_count_df = pd.DataFrame({"Shooting Count": age_shooting_count})
# Sorting the dataframe by largest count first
age_shooting_count_df_sort = age_shooting_count_df.sort_values("Shooting Count", ascending = False)
# Displaying the dataframe
age_shooting_count_df_sort
```

```python
# Now resetting the index of the age column, this will help us out with our pie charts later
age_shooting_count_index = age_shooting_count_df_sort.reset_index()
age_shooting_count_index # Displaying the dataframe
```

```python
age_shooting_count_df_index = age_shooting_count_df.reset_index()
age_shooting_count_df_index
```

```python
# Now creating the code for the bar chart - This is for a different view, to show the age demographics by bar chart
sns.set() # Using seaborn
# Defining x and y axis. x Axis is the state and y axis is the Shooting count
x_axis = np.arange(len(age_shooting_count_df_index["Age Demographics"]))
y_axis = age_shooting_count_df_index["Shooting Count"]
# Tell matplotlib that we will be making a bar chart
# We apply align="edge" to ensure our bars line up with our tick marks, and we assign respective colors
plt.bar(x_axis, y_axis, color=["lightcoral", "grey", "lightskyblue", "yellow", "lightgreen", "cyan", "orange", "turquoise", "maroon"], alpha=0.5, align="edge")
# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value+0.4 for value in x_axis]
plt.xticks(tick_locations, age_shooting_count_df_index["Age Demographics"], fontsize = 14, rotation = 15)
#Give our chart Y Label and Title
plt.title("Number of Shooting Victims by Age for timeframe 2005-2018 (Charlotte Data)", fontsize = 17)
# Printing annotation text with its respective font size
plt.annotate(s='Note:\nShooting rate increases as a person gets older. It is the most for the age group 20s and 30s and then it gradually slows down',xy= (0,10), xytext=(0,550), fontsize = 14)
plt.ylabel("Shooting Count", fontsize = 14)
# Code for increasing fig size for prettier looks :)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 15 
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size
# Saving the Image
plt.savefig("Desktop/PythonStuff/Age Percentage Bar Chart.png") # saving the graph as a .png image
# Showing the plot
plt.show()
```
